```python
wage_data=visa_df['prevailing_wage']
q1=np.percentile(wage_data,25)
q2=np.quantile(wage_data,0.50)
q3=np.percentile(wage_data,75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
con1=wage_data<lb
con2=wage_data>ub
con3=con1|con2 # | its an bitwise or
len(visa_df[con3])
outliers_data=visa_df[con3]
len(outliers_data)
```
